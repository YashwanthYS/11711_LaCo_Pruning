## PIQA Benchmark

```python
import os
import requests
from zipfile import ZipFile
from io import BytesIO

def download_and_place_dataset(url, target_dir, file_name=""):
    os.makedirs(target_dir, exist_ok=True)

    if not file_name:
        file_name = url.split("/")[-1]
    file_path = os.path.join(target_dir, file_name)

    if os.path.exists(file_path) or (file_name.endswith(".zip") and os.path.exists(file_path.replace(".zip", ""))):
        print(f"{file_name} already exists in the target directory.")
        return

    print(f"Downloading {file_name} from {url}...")
    response = requests.get(url)
    response.raise_for_status()

    if file_name.endswith(".zip"):
        with ZipFile(BytesIO(response.content)) as zip_file:
            zip_file.extractall(target_dir)
            print(f"Extracted ZIP file to {target_dir}.")
    else:
        with open(file_path, "wb") as f:
            f.write(response.content)
        print(f"Downloaded and saved to {file_path}.")

url = "https://storage.googleapis.com/ai2-mosaic/public/physicaliqa/physicaliqa-train-dev.zip"
target_dir = "/content/opencompass/data/"
download_and_place_dataset(url, target_dir)
```
